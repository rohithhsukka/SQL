**SQL Statements or Queries**
  * DDL (Data Definition Language)
    * it is used to define, manage and modify structure of database. These statements allows us to create , alter , drop database objects such as tables, indexes and views.
      * create database
      * create table
      * define column
      * set constraints

      **queries**
      * create database cx_db;
      * create table new_students (student_id int primary key, name varchar(20) , email varchar(30) unique, age int not null);
      * alter table new_students add column domain char(20);
      * alter table new_students modify name varchar(40);
      * alter table new_students drop column domain;
      * drop table students;


  * DQL (Data Query Language)
    * It is used to query data from data table
      * **queries**
      * select * from new_students;
      * select name,email from new_students;
      * select * from new_students where student_id = 3;
      * select * from new_students order by name;
      * select * from new_students order by name,age;
      * select * from new_students order by age;
      * select * from new_students order by age desc;


  * DML (Data Manipulation Language)
    * It is used for managing data within a data base.
    * **queries**
      * insert into new_students (student_id,name,age) values (4,'abhay',19);
      * update new_students set email = 'ajay@abc.com' where name = 'ajay';
      * delete from new_students where student_id = 4;
      * select * from new_students where age > 18;
